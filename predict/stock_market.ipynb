```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/stock_data.csv')
df.head()
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
print(df.shape)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

features = ['Open', 'Close','High', 'Low','Volume','RSI','MACD','Bollinger_Upper','Bollinger_Lower','Sentiment_Score','GDP_Growth','Inflation_Rate']
X = df[features]
y = df['Target']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Train-Test Spit

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print(f"Training set shape: {X_train.shape}")
print(f"Test set shape: {X_test.shape}")
```

Building the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam

model = Sequential()

model.add(LSTM(50, activation='tanh', input_shape=(X_train.shape[1], 1)))

model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Reshape input data for LSTM (3D: samples, timesteps, features)
X_train_reshaped = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test_reshaped = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Train the model
model.fit(X_train_reshaped, y_train, epochs=50, batch_size=32)
```

Model Evaluation

```python
loss, accuracy = model.evaluate(X_test_reshaped, y_test)
print(f"Test Loss: {loss}")
print(f"Test Accuracy: {accuracy}")
```

XGBoost for Residual Correction

```python
import xgboost as xgb

y_train_residuals = y_train - model.predict(X_train_reshaped).flatten()
y_test_residuals = y_test - model.predict(X_test_reshaped).flatten()

# Train an XGBoost model on the residuals
xgboost_model = xgb.XGBRegressor(objective='reg:squarederror', learning_rate=0.1)

# Train the model on the residuals
xgboost_model.fit(X_train, y_train_residuals)

# Correct the residuals for the test set
y_pred_residuals = xgboost_model.predict(X_test)

# Combine the LSTM and XGBoost predictions (add residual corrections)
final_predictions = model.predict(X_test_reshaped).flatten() + y_pred_residuals
```

Model Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report

# Convert final predictions into binary classification (Buy/Sell)
final_predictions_binary = (final_predictions > 0.5).astype(int)

# Evaluate the combined model
accuracy = accuracy_score(y_test, final_predictions_binary)
print(f"Final Model Accuracy: {accuracy}")
print(classification_report(y_test, final_predictions_binary))
```

Strengths
- Excellent at detecting "Sell" signals (99% recall)
- Very reliable "Sell" predictions (97% precision)
- High overall accuracy (96.85%)

Weaknesses
- Misses many "Buy" opportunities (only 55% recall)
- "Buy" predictions need improvement (67% F1-score)

Class Imbalance Issue:
- The model performs much better on "Sell" predictions because there are likely more "Sell" examples in our training data.
